```python
client_countries = set(clients['country'].unique())
mapping_codes = set(countries['alpha-2'].unique())
print(client_countries - mapping_codes)
```

```python
# Amend KP name to South Korea
countries.loc[countries['alpha-2'] == 'KP', 'name'] = "Korea (Democratic People's Republic of)"

# Add UK(for reference ISO code GB)
country_lookup = countries.set_index('alpha-2')[['name', 'region', 'sub-region']].to_dict('index')
country_lookup['UK'] = {'name': 'United Kingdom', 'region': 'Europe', 'sub-region': 'Northern Europe'}
```

```python
clients['region'] = clients['country'].map(lambda c: country_lookup[c]['region'])
clients['sub_region'] = clients['country'].map(lambda c: country_lookup[c]['sub-region'])

print(clients['region'].isnull().sum())
```
